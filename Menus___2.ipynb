```python
# para realizar las cabeceras, en vez de ir realizando una a una y ajustar
# indico el texto, ejecuto y solo tengo que copiar y pegar

ancho = 100


def centrar_con_asteriscos(texto, ancho):
  """Centra un texto dentro de un ancho espec칤fico, rode치ndolo de asteriscos.

  Args:
    texto: El texto a centrar.
    ancho: El ancho total de la l칤nea, incluyendo los asteriscos.

  Returns:
    Una cadena con el texto centrado y rodeado de asteriscos.
  """

  espacios = (ancho - len(texto) - 10) // 2
  return "*" * (espacios + 4) + " " + texto + " " + "*" * (espacios + 4)

# Ejemplo de uso:
texto = "MENU 51 TE ADIVINO TU NUMERO DEL 1 AL 100"
resultado = centrar_con_asteriscos(texto, ancho)
print(resultado)
```

```python
# empezamos instalando las librerias necesarias
!pip install tqdm

# despues debemos de instalar de las librerias lo necesario
from tqdm import tqdm
import time
import random

def juego_adivinar_numero():
    # vamos a realizar un juego de adivinar numeros
    # con cada intento debes esperar un tiempo que se ir치 INCREMENTANDO
    # se ver치 una barra de calaveras en la espera



    # Inicializo variables ***************************************************

    jugar_inicio = "S"
    jugar = "S"

    # defino funciones *******************************************************

    def esperar():
        print(f"{nombre} vas a tener que esperar {tiempo_espera / 10} segundos")
        for i in tqdm(range(tiempo_espera), bar_format='{l_bar}{bar}', ascii= " " + "游"):
            time.sleep(0.1)
        print()

    def desviacion():
        # calculo la desviacion del numero introducido con el esperado
        desviacion = abs(suposicion - numero_aleatorio) / abs(numero_aleatorio) * 100
        # Mostrar el resultado
        print(f"El n칰mero {suposicion} est치 alejado en un {round(desviacion, 4)}%")

    def verificar_numero(a):
        # verifico que el numero que se introduce esta entre el 1 y el 100
        if a < 1:
                a = 1
        elif a > 100:
                a = 100
        else:
            a = a
        return a



    # Saludo y solicitud de juego
    nombre = input("Hola, dime como te llamas: ").capitalize()
    jugar_inicio = input(f"{nombre}, quieres jugar a un juego de adivinanza: (S/N): ").upper()

    # aqui empieza la primera parte del juego, si gano y quiero seguir jugando salgo del desarrollo del juego sigo por aqui
    while jugar_inicio == "S":

        try:



            suposicion = int(input(f"{nombre}, dime un numero del 1 al 100: "))
            suposicion = verificar_numero(suposicion)
            numero_aleatorio = random.randint(1, 100)       # Generar un n칰mero entero aleatorio entre 1 y 100 (incluidos)
            print(numero_aleatorio)                         # esto es trampa, hay que borrar esta linea
            tiempo_espera = 0                               # estas variables las inizializo para cada vez que retome el juego
            contador = 0
            num_mayor = 100
            num_menor = 1
            jugar = "S"

                # aqui esta el desarrollo del juego

            while jugar == "S":
                contador += 1                       # cuento los intentos
                tiempo_espera += 50                 # incremento la espera 5 segundos

                if suposicion > numero_aleatorio:
                    print("El numero que has elejido es MAYOR")
                    num_mayor = suposicion
                    desviacion()
                    esperar()
                    suposicion = int(input(f"{nombre}, dime otro numero entre {num_menor} y {num_mayor}: "))
                    suposicion = verificar_numero(suposicion)

                elif suposicion < numero_aleatorio:
                    print("El numero que has elejido es MENOR")
                    num_menor = suposicion
                    desviacion()
                    esperar()
                    suposicion = int(input(f"{nombre}, dime otro numero entre {num_menor} y {num_mayor}: "))
                    suposicion = verificar_numero(suposicion)

                elif suposicion == numero_aleatorio:
                    print()
                    print(f"ENHORABUENA {nombre}, 游꿀游꿀游꿀游꿀游꿀 HAS GANADO EN {contador} INTENTOS")
                    print()
                    jugar = "N"     # con esto ya salgo del desarrollo de la partida

            jugar_inicio = input(f"{nombre}, quieres seguir jugando: (S/N): ").upper()
            print()

                # cuando he ganado pregunto si quiero seguir
                # SI, voy al primer while ----> jugar_inicio    inicio del juego
                # NO, salgo de los dos while


        except Exception as e:
        # Captura cualquier excepci칩n y guarda los detalles en 'e'
            print(f"Ocurri칩 un error, COMIENZA DE NUEVO: {e}")

        finally:
            print()

    print(f"Hasta la vista {nombre} ")
    print()
    print(f"**************************** MENU 50 ME ADIVINAS UN NUMERO DEL 1 AL 100 ****************************")
    print()
```
